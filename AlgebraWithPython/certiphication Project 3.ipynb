# Scatter plot game

Randomly generate points on a graph and the player has to input the (x,y) coordinates
For added difficulty, make the graph larger

```python
import random as rnd
import re
import math
import matplotlib.pyplot as plt

def displaying_axis(xmin:int, xmax:int, ymin:int,ymax:int):
    
  fig, ax = plt.subplots()
  ax.grid(True)
  ax.set_xlabel("x")
  ax.set_ylabel("y")
  plt.axis([xmin,xmax,ymin,ymax]) # * window size
  plt.plot([xmin,xmax],[0,0],'black') # * blue x axis
  plt.plot([0,0],[ymin,ymax], 'black') # * blue y axis

  return ax

print('The first  to reach 50 points wins. It\'s you against the CPU.')
xmin,ymin,xmax,ymax= -40,-40,40,40 
result = {'player':0,'cpu':0} 
    
while result.get('player',0)<50 and result.get('cpu',0)<50:
    ax=displaying_axis(xmin, xmax, ymin,ymax)
    x_rnd = rnd.randint(xmin,xmax) 
    y_rnd = rnd.randint(ymin,ymax) 
    plt.plot(x_rnd,y_rnd,'ro')
    ax.set_title(f'Score => Player: {result.get('player')} - CPU: {result.get('cpu')}')
    plt.show()
    coord= input('Guess ! Enter (x,y) coordinate : ')
    xy = coord.split(',')
       
    if len(xy) !=2  or not re.search('^[(]-?[0-9]+,-?[0-9]+[)]$',coord):
        print('Wrong coordinate')   
        continue
    else:
        try:
            x,y = ( int(str(xy[0])[1:]), int( str(xy[1])[:len(str(xy[1]))-1] ) )
        except:
            print('Wrong coordinate. Format is (int, int) ')
            continue
    if x_rnd==x and y_rnd==y:
        print('Good !!!')
        result['player']=result.get('player')+10
    else:
        print(f'Too bad, the correct coordinate was ({x_rnd},{y_rnd})')
        result['cpu']=result.get('cpu')+10

print(result)
if result.get('player')==50:
    print('Congrats !, Player WIN ')
else:
    print('Try again !, CPU WIN ')

```

# Algebra practice game



Generate one-step and two-step problems with random integer values and the player has to input the answer
Use positive and negative values. For added difficulty, make the numbers larger

```python
import random as rnd

# * Simplest one-step addition
def one_step_add_substract(a,b):
    print("x + ", a, " = ", b)
    return (b - a, b + a*-1)[a > 0]

# * One-step multiply
def one_step_mult(a,b):
    print(a, " * x = ", b)
    return round(b/a,2)

# * One-step divide
def one_step_div(a,b):
    print("x / ", a, " = ", b)
    return b*a

# * Two-step problems
def two_step(a,b,c):
    print(a, " * x + ", b, " = ", c)
    return round((c-b)/a,2)

answer = 0

print('The first  to reach 50 points wins. It\'s you against the CPU.')
result = {'player':0,'cpu':0} 
    
while result.get('player',0)<50 and result.get('cpu',0)<50:
    a = rnd.randint(-100,100)
    b = rnd.randint(-100,100)
    c = rnd.randint(-100,100)
    option = rnd.randint(1,5)
    
    try:
        if option==1:
            answer = one_step_add_substract(a,b)
        elif option==2:
            answer = one_step_mult(a,b)
        elif option==3:
            answer = one_step_div(a,b)
            print("Round your answer to two decimal places.")
        else:
            answer = two_step(a,b,c)
            print("Round your answer to two decimal places.")
        
        ans = input("x = ")
        if float(ans) == float(answer):
            print("Correct!")
            result['player']=result.get('player')+10
        else:
            print("Try again, The correct answer was ", answer)
            result['cpu']=result.get('cpu')+10
    except:
        print('Operation Error')
        continue
    finally:
        print(result,'\n')


if result.get('player')==50:
    print('Congrats !, Player WIN ')
else:
    print('Try again !, CPU WIN ')
```

# Projectile game - Lvl 1

Display a "wall" with random height and location. Player has to move sliders to adjust a parabolic path to clear the wall
For added difficulty, make a second level where players enter a, b, and c without sliders

```python
%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np
import math


def f(a=-5,b=20,c=0):
    print(f'A toy rocket has to clear a wall {XWALL} meters away')
    print(f'Height of the wall : {HWALL} meters')
    xmin, xmax, ymin,ymax = (0,100,0,1000)
    plt.axis([xmin,xmax,ymin,ymax]) # window size
    plt.plot([xmin,xmax],[0,0],'k') # blue x axis
    plt.plot([0,0],[ymin,ymax], 'k') # blue y axis
    
     # * vertex
    vx = - b /(2*a)
    vy = a * vx ** 2 + b * vx + c
    root = round((-b - math.sqrt(b**2-4*a*c))/(2*a),2)
    
    # * plot wall
    plt.plot([XWALL,XWALL],[0,HWALL],'b', linewidth=5)
    
    # * crash ?
    crash = a * XWALL ** 2 + b * XWALL + c
    #print(crash)
    
    result= vy>HWALL and root>XWALL and vy<ymax and crash > HWALL
    
    x = np.linspace(xmin, xmax, 1000)
    y= a*x**2 + b*x + c
    plt.title( f'{('You are Losing','You are winning')[result]} {round(a,2)}*x**2+{b}*x+{c}')
    
    # Change the next line to add c to the end of the function:
    plt.plot(x,y,('r','g')[result])
    plt.show()

# * Build wall
XWALL = rnd.randint(20, 100) 
HWALL = rnd.randint(400, 1000)

# * Change the next line to add a slider to change the c value
interactive_plot = interactive(f, a=(-5, -0.1,0.1), b=(20,250,1),c=(0,int(HWALL/4)))
interactive_plot


```
